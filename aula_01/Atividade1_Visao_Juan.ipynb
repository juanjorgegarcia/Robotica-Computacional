# Equalização de histograma

```python
rintin = cv2.imread("RinTinTin.jpg")
```

```python
rintin.shape
```

```python
rintin_gray = cv2.cvtColor(rintin, cv2.COLOR_RGB2GRAY)
```

```python
rintin_gray.shape
```

```python
plt.imshow(rintin_gray, cmap="Greys_r", vmin=0, vmax=255)
```

```python

hist,bins = np.histogram(rintin_gray.flatten(),256,[0,256])

cdf = hist.cumsum()
cdf_normalized = cdf * hist.max()/ cdf.max()

plt.plot(cdf_normalized, color = 'b')
plt.hist(rintin_gray.flatten(),256,[0,256], color = 'r')
plt.xlim([0,256])
plt.legend(('cdf','histogram'), loc = 'upper left')
plt.show()
```

## 2.1) Não é possivel identificar a caixa vermelha e a bateria azul somente analisando os histogramas e os canais R e B separadamente.

```python
hall = cv2.imread("hall_box_battery1.jpg")
hall_rgb = cv2.cvtColor(hall, cv2.COLOR_BGR2RGB)
hall_red = hall_rgb[:,:,0]

hist,bins = np.histogram(hall_red.flatten(),256,[0,256])

cdf = hist.cumsum()
cdf_normalized = cdf * hist.max()/ cdf.max()

fig = plt.figure(figsize=[10,7])
ax1 = fig.add_subplot(2,2,1)
ax1.plot(cdf_normalized, color = 'b')
plt.title("Red Pixel histogram")
ax1.hist(hall_red.flatten(),256,[0,256], color = 'r')
plt.xlim([0,256])
ax1.legend(('cdf','histogram'), loc = 'upper left')
plt.show()

hall = cv2.imread("hall_box_battery1.jpg")
hall_rgb = cv2.cvtColor(hall, cv2.COLOR_BGR2RGB)
hall_blue = hall_rgb[:,:,0]

hist,bins = np.histogram(hall_blue.flatten(),256,[0,256])

cdf = hist.cumsum()
cdf_normalized = cdf * hist.max()/ cdf.max()

fig = plt.figure(figsize=[10,7])
ax1 = fig.add_subplot(2,2,1)
ax1.plot(cdf_normalized, color = 'b')
plt.title("blue Pixel histogram")
ax1.hist(hall_blue.flatten(),256,[0,256], color = 'r')
plt.xlim([0,256])
ax1.legend(('cdf','histogram'), loc = 'upper left')
plt.show()


```

## 2.2)Isolando as figuras.

```python
hsv = cv2.cvtColor(hall, cv2.COLOR_BGR2HSV)

red_min=np.array([0, 100, 20],np.uint8)
red_max=np.array([10, 255, 255],np.uint8)

mask_red = cv2.inRange(hsv,red_min,red_max )
res = cv2.bitwise_and(hall_rgb,hall_rgb, mask = mask_red)

fig = plt.figure(figsize=[10,7])
ax1 = fig.add_subplot(2,2,1)
ax1.imshow(hall_rgb, cmap="Greys_r", vmin=0, vmax=255)
plt.title("Foto normal")
ax2 = fig.add_subplot(2,2,2)
ax2.imshow(mask_red, cmap="Greys_r", vmin=0, vmax=255)
plt.title("Caixa vermelha isolada")
ax3 = fig.add_subplot(2,2,3)
ax3.imshow(res, cmap="Greys_r", vmin=0, vmax=255)
plt.show()
```

```python
blue_min=np.array([80, 130, 50],np.uint8)
blue_max=np.array([100, 255, 255],np.uint8)

mask_blue = cv2.inRange(hsv,blue_min,blue_max )


fig2 = plt.figure(figsize=[10,7])
ax1 = fig2.add_subplot(2,2,1)
ax1.imshow(hall_rgb, cmap="Greys_r", vmin=0, vmax=255)
plt.title("Foto normal")
ax2 = fig2.add_subplot(2,2,2)
plt.title("Bateria 'azul' isolada")
ax2.imshow(mask_blue, cmap="Greys_r", vmin=0, vmax=255)
plt.show()

```
